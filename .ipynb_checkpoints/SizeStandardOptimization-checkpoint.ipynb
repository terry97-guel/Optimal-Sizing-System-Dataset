```python
Ns = [1,2,3,4,5,6,7]  # 사이즈 규격 개수
C1 = 1
C2 = 0.075
max_epoch = 100 # 최적화 횟수

# 최적화에 사용될 치수 지정
write_measure_names = ['392. 손직선길이 ','413. 손안쪽가쪽직선길이 ']
write_measure_names_eg = ["Hand Length", "Hand Width"]

# Plot
marker_size = 10
```

# Preprocessing Data

```python
!git clone https://github.com/terry97-guel/Hands-Dataset.git
```

```python
# 데이터 읽기
import pandas as pd

rawData = pd.read_csv('Hands-Dataset/SizeData.csv')
```

```python
# 남자 데이터만 추출
Data = rawData[rawData['성별']=='남']
```

```python
# 원하는 치수만 추출
measure_names = []
for col in Data.columns.tolist():
    for name in write_measure_names:
        if name in col: measure_names.append(col)

Data = Data.loc[:,measure_names]

# Delete rows with nans
temp = Data.isna()
Data = Data[~temp.any(axis=1)]

# Check 
Data.isna().any()
Data = Data.reset_index(drop=True)
Data
```

```python
import torch

import pandas as pd
import numpy as np
TotalData = Data.to_numpy()

TotalNumber = len(TotalData)
TrainData = TotalData[:int(TotalNumber*0.9)]
ValData =  TotalData[int(TotalNumber*0.9):]


print(TrainData)
```

```python
xmin = np.min(TrainData,axis=0)
xmax = np.max(TrainData,axis=0)

print(xmin, xmax)
```

# Optimize

```python
objective_N = []
u_opts_N = []
for N in Ns:
    # 최소,최대 수치를 N 등분 함.
    x_det = np.linspace(xmin,xmax,N+1)

    # 생산규격 치수의 최적화 이전 초기 예상 값들
    temp = np.zeros_like(x_det)
    temp[:-1] = x_det[1:]

    u = ((x_det+temp)/2)[:-1]
    
    
    # 최적화 시작값
    x_det_opt = torch.tensor(x_det[1:-1])
    x_det_opt.requires_grad = True

    u_opt = torch.tensor(u)
    u_opt.requires_grad = True

    # Traindata에서 최적화
    from torch import optim
    from matplotlib import pyplot as plt
    x = torch.tensor(TrainData)
    def getLossFn(x):
        def TargetLoss(u_opt):
            x_dist = x[:,0].unsqueeze(0) - u_opt[:,0].unsqueeze(1)
            
            # 수치1에 대한 손실값
            loss_matrix1 = torch.where(
                x_dist>0,
                C2*torch.nn.MSELoss(reduction='none')(x_dist,torch.zeros_like(x_dist)),
                C1*torch.nn.MSELoss(reduction='none')(x_dist,torch.zeros_like(x_dist))
                )
            
            # 수치2에 대한 손실값
            x_dist = x[:,1].unsqueeze(0) - u_opt[:,1].unsqueeze(1)
            loss_matrix2 = torch.where(
                x_dist>0,
                C2*torch.nn.MSELoss(reduction='none')(x_dist,torch.zeros_like(x_dist)),
                C1*torch.nn.MSELoss(reduction='none')(x_dist,torch.zeros_like(x_dist))
                )

            # 전체 손실값
            loss_matrix = loss_matrix1 + loss_matrix2        

            # 전체 손실값 중 최적을 사용자가 선택
            val,_ = torch.min(loss_matrix,dim=0)
            return torch.mean(val)
        return TargetLoss


    f = getLossFn(x)
    h_lbfgs = []
    u_opts = []
    
    optimizer = optim.LBFGS([u_opt],
                history_size=1,
                max_iter=20,
                lr=0.1,
                line_search_fn="strong_wolfe")

    for epoch in range(max_epoch):
        optimizer.zero_grad()
        objective = f(u_opt)
        objective.backward()
        optimizer.step(lambda: f(u_opt))
        h_lbfgs.append(objective.item())
        u_opts.append(u_opt.detach().clone())
        
        # print(objective)
        print(f"[Epoch{epoch+1}]Optimal Value of Production Specification \n", u_opt)
        print("Loss:{:.4f}".format(objective.clone().detach()))
        
    objective_N.append(h_lbfgs)
    u_opts_N.append(u_opt)
    optimizer = optim.LBFGS([u_opt],
                history_size=1,
                max_iter=20,
                line_search_fn="strong_wolfe")
                
    plt.plot(h_lbfgs, label=f'N = {N}', color = 'k')
    plt.xlabel("Optimization Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
```

```python
# 형용규격

ID_standard = [
    175.0, 82.0,
    180.0, 87.0,
    185.0, 92.0,
    190.0, 97.0,]
ID_standard = torch.tensor(ID_standard).reshape(-1,2)

ID_standard_loss = f(ID_standard)
```

```python
for N,objetives in zip(Ns,objective_N):
    plt.plot(objetives,"-", label = f"N= {N}")
    plt.xlabel("Optimization Epoch")
    plt.ylabel("Loss")

plt.plot(
    [ID_standard_loss]*max_epoch,"--", label="ID Standard")
plt.legend()
```

```python
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


def plot_colortable(colors, sort_colors=True, emptycols=0):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(color))),
                         name)
                        for name, color in colors.items())
        names = [name for hsv, name in by_hsv]
    else:
        names = list(colors)

    n = len(names)
    ncols = 4 - emptycols
    nrows = n // ncols + int(n % ncols > 0)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig
```

```python
# ValData에서 최적화 결과 검증
from torch import optim
from matplotlib import pyplot as plt


def ValidationLoss(u_opt):
    x = torch.tensor(ValData)
    x_dist = x[:,0].unsqueeze(0) - u_opt[:,0].unsqueeze(1)
    loss_matrix1 = torch.where(
        x_dist>0,
        C2*torch.nn.MSELoss(reduction='none')(x_dist,torch.zeros_like(x_dist)),
        C1*torch.nn.MSELoss(reduction='none')(x_dist,torch.zeros_like(x_dist))
        )


    x_dist = x[:,1].unsqueeze(0) - u_opt[:,1].unsqueeze(1)
    loss_matrix2 = torch.where(
        x_dist>0,
        C2*torch.nn.MSELoss(reduction='none')(x_dist,torch.zeros_like(x_dist)),
        C1*torch.nn.MSELoss(reduction='none')(x_dist,torch.zeros_like(x_dist))
        )

    loss_matrix = loss_matrix1 + loss_matrix2        

    val,_ = torch.min(loss_matrix,dim=0)
    return torch.mean(val)


valLoss = []
for N,u_opt in zip(Ns,u_opts_N):
    f = ValidationLoss(u_opt)
    valLoss.append(f.detach().cpu())


plt.plot(Ns,valLoss)
plt.ylabel("Validation Loss")
plt.xlabel("N")
plt.legend()
plt.show()
```
